Exp 5: Study of dictionary in python


```python
# creation of dictionary
dict = {
    "brand": "ford",
    "model": "mustang",
    "year": 1964
}
print(dict)
```

```python
# RETRIVING VALUES FROM DICTIONARY
dict = {
    "brand": "ford",
    "model": "mustang",
    "year": 1964
}
print(dict["brand"])
```

```python
# Dictionary takes latest value given for key. It will not allow duplicates
dict = {
    "brand": "ford",
    "model": "mustang",
    "year": 1964,
    "year": 2020
}
print(dict)

# length of key-value pairs
len(dict)
```

```python
# changing or adding key:value in dictionary
car = {
    "brand": "ford",
    "model": "mustang",
    "year": 1964,
    "year": 2020
}
car["color"] = "red"
print(car)
x = car.values()
print(x) # after the change
print(type(car))
```

```python
# retrival of values
student = {
    "roll": 74,
    "name": "medhansh",
    "branch": "E&TC"
}
student.pop("branch")
print("after removal: ",student)
```

```python
'''1. A dictionary stored product names and their prices.
Write a python program to update the price of a given product.'''
product = {
    "pen": 10,
    "book": 50,
    "scale": 20
}
print("original prices: ",product)
product["book"] = 55
print("updated prices: ",product)
```

```python
'''2. A dictionary stores student names and marks.
Write a python program to search for a students marks.
If the student is not found, display a suitable message.'''
student = {
    "a": 80,
    "b": 90,
    "c": 70
}
name = input("enter student name: ")
print(student.get(name, "student not found"))
```

```python
'''3. A dictionary stores usernames and passwords.
Write a python program to validate user login.'''
users = {"admin": '1234', "student": "abcd"}
username = input("enter username: ")
password = input("enter password: ")
if users.get(username) == password:
    print("login successful")
else:
    print("invalid username or password")
```

```python
'''4. A dictionary stores student name and marks.
Write a python program to find the student who scored the highest marks the highest marks.'''
marks = {"a":85, "b":92, "c" : 78}
topper = max(marks, key=marks.get)
print("topper: ",topper)
print("marks: ",marks[topper])
```
